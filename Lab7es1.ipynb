```
import numpy as np
import tensorflow as tf
print(tf.__version__) # Has to be 2.2
n=10

train_loss_results = []
train_accuracy_results = []

validation_loss_results = []
validation_accuracy_results = []


# Example modified from: https://keras.io/api/models/model/
class LinearRegression(tf.keras.Model):
    def __init__(self):
        super(LinearRegression, self).__init__()
        self.parameters=tf.Variable((0.0,0.0))          
        self.dense1 = tf.keras.layers.Dense(1, activation='relu')

    def compile(self, optimizer, loss, metric):
        """
        Overriden .compile() method to initialize
        optimizer, loss, and metric for train_step function
        """
        super(LinearRegression,self).compile()
        self.optimizer = optimizer
        self.loss=loss
        self.metric = metric
        #self.epochs=100

        #model.add_loss()
    
    def train_step(self, data):
        print(f"Eager execution mode: {tf.executing_eagerly()}")
        X, y = data
        #print(X)       
        # Track gradients
        with tf.GradientTape() as tape:
            tape.watch(self.parameters)
            y_pred = self(X)#, training=True)
            #print(type(y_pred))
            loss = self.loss(y, y_pred)
             

        # Compute and modify the weights
        grads = tape.gradient(loss, self.trainable_variables)
        self.optimizer.apply_gradients(zip(grads, self.trainable_variables))  
        #print(grads)      

        self.metric.update_state(y, y_pred)
        # Return a dict mapping metric names to current value
        return {m.name: m.result() for m in self.metrics}

        # Compute metric
        #metric = self.metric(y, y_pred)
        #return {"loss": loss, "metric": metric} 

    def call(self, inputs):        
        return self.parameters[0]*X+self.parameters[1]#self.dense1(inputs)
```

```
X = tf.random.uniform([10,2])
noise = tf.random.normal([10, 2])
y = X  + 2 + noise
```

```
tf._api.v2.config.run_functions_eagerly(True)
print(f"Eager execution mode: {tf.executing_eagerly()}")

model = LinearRegression()
model.compile(tf.compat.v1.train.GradientDescentOptimizer(learning_rate=0.02), 
                  tf.keras.losses.MeanSquaredError(),               
                tf.keras.metrics.MeanSquaredError())
    
history=model.fit(X, y,epochs=100,callbacks=[tf.keras.callbacks.History()])
print(history.history)
print(model.parameters)
```
